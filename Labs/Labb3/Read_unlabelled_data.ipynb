```python
import numpy as np
import matplotlib.pyplot as plt

filename = "unlabelled_data.csv"
data = np.genfromtxt(filename, delimiter=',') # Read data from a CSV file and separate values by commas
x = data[:, 0]
y = data[:, 1]

plt.figure(figsize=(8, 5))
plt.scatter(x, y, color="blue", alpha=0.6)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Unlabeled data")

k = -2
m = 0

x_line = np.linspace(min(x), max(x), 100)
y_line = k * x_line + m
plt.plot(x_line, y_line, color="red", label=f"y = {k}x + {m}")
plt.legend()

```
